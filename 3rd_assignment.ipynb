# extract the information needed from the big json file  


I didn't upload the original tweets file,bucause it's too huge,and I extracted only the information I need into another json file,called "tweets1" and "tweets2"

```python

f = open("sample_2.json",'r',encoding='utf-8')
ln = 0
for line in f.readlines():
    ln += 1
    dic = json.loads(line)
    with codecs.open('tweets2.json','a', 'utf-8') as outf:
        
        if 'media' in dic['entities']:
            json.dump({'id': dic['_id'], 'hashtags':dic['entities']['hashtags'] ,'urls':dic['entities']['urls'],'media':dic['entities']['media']},outf, ensure_ascii=False)
        else:
            json.dump({'id': dic['_id'], 'hashtags':dic['entities']['hashtags'] ,'urls':dic['entities']['urls'],'media':[]},outf, ensure_ascii=False)
        outf.write('\n')
f.close()
```

```python
f1 = open("sample_1.json",'r',encoding='utf-8')
ln1 = 0
for line in f1.readlines():
    ln1 += 1
    dic = json.loads(line)
    with codecs.open('tweets1.json','a', 'utf-8') as outf:
        
        if 'media' in dic['entities']:
            json.dump({'id': dic['_id'], 'hashtags':dic['entities']['hashtags'] ,'urls':dic['entities']['urls'],'media':dic['entities']['media']},outf, ensure_ascii=False)
        else:
            json.dump({'id': dic['_id'], 'hashtags':dic['entities']['hashtags'] ,'urls':dic['entities']['urls'],'media':[]},outf, ensure_ascii=False)
        outf.write('\n')
f1.close()
```

## For sample_1 

```python

f = open("tweets1.json",'r',encoding='utf-8')
ln = 0
sp1_urls = 0
sp1_media = 0
sp1_hash1 = 0
sp1_hash2 = 0
sp1_hashlist1 = []
sp1_hashlist2 = []


for line in f.readlines():
    ln += 1
    hlist = []
    dic = json.loads(line)
    
    if dic['urls'] != []:
        sp1_urls += 1
    if dic['media'] != []:
        sp1_media += 1
    if dic['hashtags'] !=[]:
        sp1_hash1 += 1
        for i in dic['hashtags']:
            sp1_hashlist1.append(i['text'])
        if len(dic['hashtags']) >= 2:
            #print(dic['id'])
            for j in dic['hashtags']:
                hlist.append(j['text'])
            sp1_hashlist2.append(hlist)
```

### Hashtags,Urls,Media

```python
print("Total number of tweets in sample_1: ",ln )
print("Total number of tweets contain urls in sample_1: ",sp1_urls,"  pertcentage: ",sp1_urls/ln  )
print("Total number of tweets contain media in sample_1: ",sp1_media,"  pertcentage: ",sp1_media/ln  )
print("Total number of tweets contain at least 1 hashtag in sample_1: ",sp1_hash1,"  pertcentage: ",sp1_hash1/ln  )
print("Total number of tweets contain more than 2 hashtags in sample_1: ",len(sp1_hashlist2),"  pertcentage: ",len(sp1_hashlist2)/ln  )
```

### The 50 most frequent hashtags pairs in sample_1

```python
pairs_s1 =[]


for i in sp1_hashlist2:
    
    for m in i:
        for j in i:
            templist = []
            if j != m:
                templist.append(j)
                templist.append(m)
                pairs_s1.append(templist)
```

```python
pairlist_s1 = []
for i in pairs_s1:
    if (i not in pairlist_s1) & (i[::-1] not in pairlist_s1):
        pairlist_s1.append(i)
```

```python
count_pair1 =[]
for i in pairlist_s1:
    templist = []
    templist.append(i)
    templist.append(pairs_s1.count(i))
    count_pair1.append(templist)
```

```python
columnsh = ["hashtags ","numbers"]
count_pdf1 = pd.DataFrame(count_pair1,columns = columnsh)
df1_2 = count_pdf1.sort_values(by="numbers" , ascending=False).head(50)
df1_2.index=[i for i in range(df1_2.shape[0])]
```
